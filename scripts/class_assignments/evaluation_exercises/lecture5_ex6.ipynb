```python
from si.io.csv import read_csv
from si.io.data_file import read_data_file
from sklearn.preprocessing import StandardScaler
from si.model_selection.split import train_test_split
```

# Ridge Regression

```python
cpu_path = r"C:\Users\anaca\Documents\GitHub\SIB-ML-Portfolio\datasets\cpu.csv"
cpu_dataset = read_csv(cpu_path, sep=",", features=True, label=True)
```

```python
cpu_dataset.x = StandardScaler().fit_transform(cpu_dataset.x)
```

```python
cpu_train, cpu_test = train_test_split(cpu_dataset)
```

```python
from si.linear_module.ridge_regression import RidgeRegression

ridge = RidgeRegression(max_iter=2000)
ridge.fit(cpu_train, gradient_descent_algorithm='static_alpha')
```

```python
ridge.predict(cpu_test)
```

```python
ridge.score(cpu_test)
```

```python
ridge.cost(cpu_test)
```

```python
ridge.cost_history
```

```python
ridge.cost_function_plot()
```

```python
ridge.fit(cpu_train, gradient_descent_algorithm='half_alpha')
ridge.cost_function_plot()
```

# Logistic Regression

```python
breast_bin_dataset = r"C:\Users\anaca\Documents\GitHub\SIB-ML-Portfolio\datasets\breast-bin.data"
dataset = read_data_file(breast_bin_dataset, sep=",", label=True)
```

```python
dataset.x = StandardScaler().fit_transform(dataset.x)
```

```python
train, test = train_test_split(dataset)
```

```python
from si.linear_module.logistic_regression import LogisticRegression

model = LogisticRegression(max_iter=2000)
model.fit(train, gradient_descent_algorithm='static_alpha')
```

```python
model.score(test)
```

```python
model.cost(test)
```

```python
model.cost_history
```

```python
model.cost_function_plot()
```

```python
model.fit(train, gradient_descent_algorithm='half_alpha')
model.cost_function_plot()
```
