## Transformação logarítmica
Log transformation is a data transformation method in which it replaces each variable x with a log(x). When original continuous data does not follow the bell curve, we can log transform this data to make it as “normal” as possible so that the statistical analysis results from this data become more valid. In other words, the log transformation reduces or removes the skewness of our original data.

```python
float_data = data.astype(float)
log_data = np.log2(float_data)
log_data.apply(lambda x: str(x.min())+" - "+str(x.max()))
```

```python
met_log_data = log_data.iloc[:,14:17].values

plt.boxplot(met_log_data)
plt.xticks([1,2,3],met_lab)
plt.title("Distribuição dos valores dos metabolitos Asparagine, Acetate, Succinate após transformação logarítmica.")
plt.ylabel("Concentração de metabolitos na urina")
plt.show()
```

### Exemplos com testes estatísticos


```python
data_alanine = data.Alanine.copy()
alanine_control = data_alanine[meta["Muscle.loss"]=="control"]

plt.close()
plt.hist(alanine_control)
plt.title("Distribuição dos valores da concentração da Alanina nas amostras controlo.")
```

```python
alanine_sample = data_alanine[meta["Muscle.loss"]!="control"]

plt.close()
plt.hist(alanine_sample)
plt.title("Distribuição dos valores da concentração da Alanina nas amostras em estudo.")
```

```python
from scipy.stats import stats

#H0: The data at hand follows a normal distribution.
#H1: The data at hand does not follow a normal distribution.
t_test, p_value = stats.shapiro(alanine_control)
p_value
#As the p value is inferior to 0,05, we reject the null hypothesis. Therefore, the data does not follow a normal distribution.
```

```python
#H0: The data at hand follows a normal distribution.
#H1: The data at hand does not follow a normal distribution.
t_test, p_value = stats.shapiro(alanine_sample)
p_value
#As the p value is inferior to 0,05, we reject the null hypothesis. Therefore, the data does not follow a normal distribution.
```

```python
alanine_control_log = np.log2(alanine_control)
alanine_sample_log = np.log2(alanine_sample)

#H0: There isn't a statistically meaningfull difference between the means of the groups.
#H1: There is a statistically meaningfull difference between the means of the groups

t_test, p_value = stats.f_oneway(alanine_control_log, alanine_sample_log)
p_value
#As the p value is inferior to 0,05, we reject the null hypothesis. Therefore, the two groups don't have a similar mean.
```

Ex.6.1) Quais os metabolitos que têm concentrações significativamente diferentes entre os dois grupos (control e cachexic)?

```python
group_metabs = lambda x : stats.ttest_ind(x[meta['Muscle.loss']=='control'],x[meta['Muscle.loss']=='cachexic'])
# Calculates the T-test for the means of the two independent groups
# Lambda functions compose identity functions, thefeore they return their argument
# Bound variable are arguments and free variables are only referenced in the expression

# metabs_de_tt(data))
data.columns[data.apply(lambda x: group_metabs(x)[1]<0.05)]
#Applies the function to all the columns/metabolites and returns only the metabolites whose mean in the control and study group have a statistically meaningfull difference
```

Ex.7) Repita o anterior usando Testes não paramétricos - Mann-Whitney U test

## Correções para testes múltiplos

```python
group_pvalues = data.apply(lambda x: group_metabs(x)[1])
group_pvalues
```

```python
import statsmodels.stats as sm
corr_pvalues = sm.multitest.multipletests(group_pvalues, method = "fdr_bh")[1]
# multipletests -> test results and p-value correction for multiple tests
# Multiple testing corrections adjust the individual p-values derived from multiple tests to correct for occurrence the overall rate/false positives'
# method -> Method used for testing and adjustment of pvalues (fdr_bh : Benjamini/Hochberg (non-negative))
corr_pvalues
```

### Correlações

```python
data_glucose = data.Glucose
data_sucrose = data.Sucrose

plt.scatter(data_glucose, data_sucrose)
plt.show()
```

```python
#H0: There is no correlation between the two metabolites.
#H1: There is a correlation between the two variables.

corr, p_value = stats.pearsonr(data_glucose, data_sucrose)
corr

#As the p value is superior to 0,05, we accept the null hypothesis. Therefore, the two groups are correlated.
```

```python
#H0: There is no correlation between the two metabolites.
#H1: There is a correlation between the two variables.

sp_corr, p_value = stats.spearmanr(data_glucose, data_sucrose)
sp_corr

#As the p value is superior to 0,05, we accept the null hypothesis. Therefore, the two groups are correlated.
```

## Análise de variância / regressão linear

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Linear regression model
# A linear regression model establishes the relation between a dependent variable(y) and at least one independent variable(x)
lm = ols("data.Leucine~data.Valine", data).fit() # fitting the regression line to the data
anova_lm(lm) # analysis of variance table for one or more linear model fits
# df -> degrees of freedom
# F value -> ratio of between and within-group mean squares
```

```python
lm.summary()
```

```python
lm.rsquared
# How well the data fit the regression model
# A value of 0.7 is a good fit.
```
